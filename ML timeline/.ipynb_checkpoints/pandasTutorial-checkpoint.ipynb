```python
df = pd.read_csv('pokemon_data.csv')
df
```

```python
# average HP of the pokemons
df['HP'].mean()
```

```python
# median HP and Speed
df[['HP','Speed']].median()
```

```python
df['HP'].describe()
```

```python
# specific combinations of aggregating statisitcs
df.agg(
    {
        'HP':['count','mean','std'],
        'Attack':['count','mean','std']
    }
)
```

# Aggregating statistics grouped by category


```python
# df['HP'].groupby('Legendary').mean()
```

```python
df.groupby('Legendary').mean(numeric_only=True)
```

```python
df.groupby('Legendary')['HP'].mean()
```

```python
# grouping can be done with more columns
df.groupby(['Legendary','Generation'])['HP'].mean()
```

# Count number of records by category

```python
df['Generation'].value_counts()
```

```python
# count excludes the null values 
df.groupby('Legendary')['Legendary'].count()
```

```python
# size includes all the values including null values
df.groupby('Type 1')['Type 2'].size()
```

```python
df.groupby('Type 1')['Type 2'].count()
```

# How to reshape the layouts of tables


```python
# Sorting table rows
df.sort_values(by=['HP'], ascending = True).head()
```

```python
new_df = df.sort_values(by=['HP','Attack'],ascending = True).head()
```

```python
# Long to wide table format

type2 = df[df['Type 1']=='Water']
type2_sub = type2.sort_index().groupby(['Type 1']).head()
type2_sub
```

```python
# generation values next to each other
gen = df.groupby(['Generation'])['Type 1'].head()
gen
```
