# Matplotlib 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('AAPL_daily_stock_data.csv')
data['Gain/Loss'] = data['Close']-data['Open']
```

```python
data
```

```python
# Scatter plot
plt.scatter(data['Open'],data['Close'])
plt.title('ScatterPlot')

plt.xlabel('Open')
plt.ylabel('Close')

plt.show()
```

```python
data = pd.read_csv('tips.csv')

plt.scatter(data['day'],data['tip'],c = data['size'],s =data['total_bill'])

plt.title('Scatter plot')

plt.xlabel('Day')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```

```python
# Line chart

data = pd.read_csv('tips.csv')
plt.plot(data['tip'])
plt.plot(data['size'])

plt.title('Line chart')
plt.xlabel('day')
plt.ylabel('tip')

plt.show()
```

```python
# Bar plot
data = pd.read_csv('tips.csv')
plt.bar(data['day'],data['tip'])
plt.title('Bat plot')
plt.xlabel('day')
plt.ylabel('tip')

plt.show()
```

```python
# Histogram
plt.hist(data['total_bill'])
plt.title('Histogram')
plt.show()
```
