```python
import numpy as np
import matplotlib.pyplot as plt

class SimpleLinearRegression:
    def __init__(self):
        self.intercept_ = None  # Represents the y-intercept (b0)
        self.coef_ = None      # Represents the coefficient (slope) (b1)

    def fit(self, X, y):
        """
        Fits the linear regression model to the training data.

        Args:
            X (array-like): Training input features (independent variable).  Should be a 1D array or a 2D array with one column.
            y (array-like): Training target values (dependent variable).
        """

        X = np.array(X).reshape(-1, 1)  # Reshape X to a 2D array (n_samples, 1) if it's 1D
        y = np.array(y)

        n_samples = X.shape[0]

        # Calculate the mean of X and y
        x_mean = np.mean(X)
        y_mean = np.mean(y)

        # Calculate the slope (coefficient)
        numerator = np.sum((X - x_mean) * (y - y_mean))
        denominator = np.sum((X - x_mean)**2)
        self.coef_ = numerator / denominator

        # Calculate the y-intercept
        self.intercept_ = y_mean - self.coef_ * x_mean

    def predict(self, X):
        """
        Predicts the target values for new input data.

        Args:
            X (array-like): Input features for prediction.  Should be a 1D array or a 2D array with one column.

        Returns:
            array-like: Predicted target values.
        """
        X = np.array(X).reshape(-1, 1) # Reshape X if needed
        return self.intercept_ + self.coef_ * X


# Example Usage:
X = np.array([1, 2, 3, 4, 5])  # Independent variable (e.g., hours studied)
y = np.array([2, 4, 5, 4, 5])  # Dependent variable (e.g., exam scores)

# Create and fit the model
model = SimpleLinearRegression()
model.fit(X, y)

# Print the coefficients
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)

# Make predictions
X_new = np.array([6, 7])  # New data points for prediction
y_pred = model.predict(X_new)
print("Predictions:", y_pred)

# Plot the data and the regression line
plt.scatter(X, y, label="Training Data")
plt.plot(X, model.predict(X), color='red', label="Regression Line") # Plot the regression line using the fitted model
plt.scatter(X_new, y_pred, color='green', label="New Predictions") # Plot the new predictions
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Simple Linear Regression")
plt.show()
```
