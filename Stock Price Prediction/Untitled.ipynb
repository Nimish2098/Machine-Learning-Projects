```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('AAPL_daily_stock_data.csv');
data.head()
```

```python
data['Gain/Loss'] = data['Close']-data['Open']
```

```python
data.head()
```

```python
x = data[['Volume','Gain/Loss']]
y = data['Close']
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

x_scaled_data = pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled_data,y,test_size=0.2,random_state = 42)
```

```python
# creating the model and  train the model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
print(mse)
print(r2)
```
