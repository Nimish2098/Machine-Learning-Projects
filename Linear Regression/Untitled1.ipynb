# Demand and Supply dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
try:
    data = pd.read_csv('rides.csv')
except FileNotFoundError:
    print("Error:dataset not fount")
    exit()
```

```python
try:
    x = data[['Drivers Active Per Hour']] #numbers of drivers active per hour (feature)
    y = data['Riders Active Per Hour'] #number of riders active per hour (target)
except KeyError as e:
    print(f"Error:Column'{e}'not found in the dataset")
    exit()
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# make predictions on the test
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f"Mean squared error:{mse}")
print(f"R-squared:{r2}")
```
